```python
#Nạp thư viện
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#
```

```python
dataset = pd.read_csv('/content/advertising.csv')
```

```python
#xử lý timestamp
dataset['Timestamp'] = pd.to_datetime(dataset['Timestamp'])

dataset['Hour'] = dataset['Timestamp'].dt.hour
dataset['day_of_week'] = dataset['Timestamp'].dt.dayofweek
dataset['month'] = dataset['Timestamp'].dt.month

dataset.head(5)
```

```python
dataset= dataset.drop(columns=['Timestamp'])
dataset.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
#Scale dữ liệu
num_cols = [
'Daily Time Spent on Site',
'Age',
'Area Income',
'Daily Internet Usage',
'hour',
'day_of_week',
'month'
]
dataset.head(5)
```

```python
#xỬ Lý text
from sklearn.feature_extraction.text import TfidfVectorizer

TfidfVectorizer(
    max_features=1000,
    stop_words='english',
    ngram_range=(1,2)
)
```

```python
#Xử lý biến categorical
for col in ['City','Country']:
    freq = dataset[col].value_counts()
    dataset[col] = dataset[col].map(freq)
```

```python
dataset.head(5)
```

```python
X = dataset.drop(columns='Clicked on Ad')
y = dataset['Clicked on Ad']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

num_features = [
'Daily Time Spent on Site',
'Age',
'Area Income',
'Daily Internet Usage',
'Hour', # Changed 'hour' to 'Hour'
'day_of_week',
'month',
'City',
'Country'
]

preprocess = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('text', TfidfVectorizer(max_features=1000, ngram_range=(1,2)), 'Ad Topic Line'),
        ('bin', 'passthrough', ['Male'])
    ]
)

model = Pipeline([
    ('prep', preprocess),
    ('clf', LogisticRegression(penalty='l2', max_iter=1000))
])
```

```python
model.fit(X_train,y_train)
```

```python
y_prob = model.predict_proba(X_test)[:,1]
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
```
